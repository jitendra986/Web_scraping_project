```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.scrapethissite.com/pages/forms/?page_num='
```

```python
list_url = []
for i in range (1,25):
    url1=url+str(i)
    list_url.append(url1)

print(list_url)
```

```python
page = []
for item in list_url:
    page_item = requests.get(item)
    page.append(page_item)
print(page)
```

```python
soup_list=[]
for i in range(0, len(page)):
    soup = BeautifulSoup(page[i].text, 'html')
    soup_list.append(soup)
```

```python
soup1=soup_list[0]
```

```python
print(soup1.prettify())
```

# Scraping Data from a Real Website + Pandas

```python
table = soup1.find_all('table')[0]
```

```python
team_titles = table.find_all('th')
```

```python
team_titles[0]
```

```python
team_table_titles = [title.text.strip() for title in team_titles]
team_table_titles
```

```python
import pandas as pd
df = pd.DataFrame(columns= team_table_titles)
df
```

```python
#column_data is a list
column_data = []
for soup in soup_list:
    table_page = soup.find_all('table')[0]
    column_data_per_page = table_page.find_all('tr')
    column_data.append(column_data_per_page)
print(column_data[0])
```

```python
for column_data_per_page in column_data:
    for row in column_data_per_page[1:]:
        row_data = row.find_all('td')
        individual_row_data = [data.text.strip() for data in row_data]
        #print(individual_row_data)
        length = len(df)
        df.loc[length] = individual_row_data    
```
